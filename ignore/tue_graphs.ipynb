```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
# don't run again
full = pd.read_csv("whatsinyourcart/full.csv")
```

```python
color1 = '#F1D78C'
color2 = '#F6A811'
color3 = '#F46708'
color4 = '#EF727F'
color5 = '#E84846'
citrus = [color1, color2, color3, color4, color5]
sns.palplot(sns.color_palette(citrus))

fonttitle = {'fontname':'Helvetica', 'fontsize':30}
fontaxis = {'fontname':'Helvetica', 'fontsize':20}
```

```python
high_reord = full.groupby('product_name').agg({"reordered":"count"})\
                     .sort_values(by='reordered', ascending = False).reset_index()[:15]
high_reord.head()
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax = sns.barplot(high_reord['reordered'], high_reord['product_name'], palette = citrus)
ax.set_xlabel('Number of Reorders', fontdict=fontaxis)
ax.set_ylabel("Item", fontdict=fontaxis)
ax.tick_params(axis='both', which='major', labelsize=18)
plt.title('Top 15 Highest Reordered Item', fontdict=fonttitle)
plt.show();
```

```python
dow = full['order_dow'].value_counts().reset_index()
dow
```

```python
days = pd.Series(["Saturday", 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])
days
```

```python
for i in dow['index']:
    dow['index'][i] = days[i]
```

```python
dow['index'], dow['order_dow']
```

```python
    fig, ax = plt.subplots(figsize = (20, 10))
    ax = sns.barplot(dow['index'], dow['order_dow'], palette = citrus)
    ax.set_ylabel('Number of Orders', fontdict=fontaxis)
    ax.set_xlabel("Day", fontdict=fontaxis)
    plt.title('Orders by Day of Week', fontdict=fonttitle)
    ax.tick_params(axis='both', which='major', labelsize=18)
    plt.show()
    plt.savefig('dayofweek');
```

```python
hours = full['order_hour_of_day'].value_counts().reset_index().sort_values(by="index")
hours
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax = sns.barplot(hours['index'], hours['order_hour_of_day'], palette = citrus)
ax.set_ylabel('Number of Orders', fontdict=fontaxis)
ax.set_xlabel("Hours", fontdict=fontaxis)
plt.title('Orders by Hour', fontdict=fonttitle)
plt.show()
plt.savefig('hour.png');
```
